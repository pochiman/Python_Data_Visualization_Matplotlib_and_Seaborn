```python
import seaborn as sns
import pandas as pd
```

```python
diamonds = pd.read_csv("../Data/Diamonds Prices2022.csv")
```

```python
sns.regplot(x="x", y="price", data=diamonds)
```

```python
import scipy.stats as stats

stats.linregress(diamonds["carat"], diamonds["price"])
```

```python
sns.lmplot(x="carat", y="price", data=diamonds)
```

```python
# Linekws

sns.lmplot(
    x="carat", 
    y="price", 
    hue="clarity",
    data=diamonds
)
```

```python
import scipy.stats as stats

# print slope and intercept for each category

for category in diamonds["clarity"].unique():
    df_filtered = diamonds.loc[diamonds["clarity"] == category]
    print(f"{category}: {stats.linregress(df_filtered['carat'], df_filtered['price'])[0:2]}")
```

```python
sns.lmplot(
    x="carat",
    y="price",
    row="clarity",
    col="cut",
    data=diamonds
)
```
