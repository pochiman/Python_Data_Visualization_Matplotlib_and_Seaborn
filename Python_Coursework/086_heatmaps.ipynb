```python
import seaborn as sns
import pandas as pd

diamonds = pd.read_csv("../Data/Diamonds Prices2022.csv")
```

```python
diamond_pivot = diamonds.pivot_table(
    index="cut",
    columns="color",
    values={"carat": "mean"}
)

diamond_pivot.head()
```

```python
sns.heatmap(diamond_pivot.droplevel(0, axis=1))
```

```python
sns.heatmap(
    diamonds.corr(),
    annot=True,
    cmap="RdYlGn"
)
```

```python
# Select only numeric columns
numeric_diamonds = diamonds.select_dtypes(include="number")

# Create the heatmap
sns.heatmap(
    numeric_diamonds.corr(),
    annot=True,
    cmap="RdYlGn"
)
```
