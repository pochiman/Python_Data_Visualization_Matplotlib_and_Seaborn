```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
housing_raw = pd.read_csv(
    "../Data/housing_data.csv",
    parse_dates=["period_begin", "period_end"],
    dtype={"total_homes_sold": "Int32", "region_name": "category"})
```

```python
ca_housing = (
    housing_raw
    .loc[(housing_raw["region_name"].str.contains("CA"))]
    .assign(region_name = housing_raw["region_name"].str[0:-11])
    .set_index("period_begin").sort_index()
)
```

```python
ca_housing.head()
```

```python
sd_dual = (
    ca_housing
    .loc[ca_housing["region_name"] == "San Diego",
        ["inventory", "median_active_list_price"]]
)

sd_dual.head()
```

```python
fig, ax = plt.subplots()

ax.plot(sd_dual)
```

```python
fig, ax = plt.subplots()

ax.plot(sd_dual["inventory"])
```

```python
fig, ax = plt.subplots()

ax.plot(sd_dual["inventory"])
ax.plot(sd_dual["median_active_list_price"])
```

```python
fig, ax = plt.subplots()

ax.plot(sd_dual["inventory"])

ax2 = ax.twinx()

ax2.plot(sd_dual["median_active_list_price"])
```

```python
fig, ax = plt.subplots()

ax.plot(sd_dual["inventory"])

ax.legend("Inventory")

ax2 = ax.twinx()

ax2.plot(sd_dual["median_active_list_price"])

ax2.legend(["Price"])
```

```python
fig, ax = plt.subplots()

ax.plot(sd_dual["inventory"],
       label="Inventory")

ax.set_ylim(0, 10000)

ax2 = ax.twinx()

ax2.plot(sd_dual["median_active_list_price"],
        label="Price",
        c="orange")

ax2.set_ylim(0, 1000000)


fig.legend(bbox_to_anchor=(.35, .28))
```
