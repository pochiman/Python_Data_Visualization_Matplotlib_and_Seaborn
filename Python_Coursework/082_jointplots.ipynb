```python
import seaborn as sns
import pandas as pd

diamonds = pd.read_csv("../Data/Diamonds Prices2022.csv")
```

```python
sns.jointplot(x="carat", y="price", data=diamonds)
```

```python
sns.jointplot(
    x="carat", 
    y="price",
    color="orange",
    scatter_kws={'alpha':.1},
    # hue='clarity',
    kind="reg", # hist, reg, hex, scatter, kde, resid,
    # palette="viridis",
    
    data=diamonds #.query("carat <= 1 and price<6000")
)
```

```python
sns.jointplot(
    x="carat", 
    y="price",
    color="orange",
    # scatter_kws={'alpha':.1},
    # hue='clarity',
    kind="hex", # hist, reg, hex, scatter, kde, resid,
    # palette="viridis",
    
    data=diamonds.query("carat <= 1 and price<6000")
)
```

```python
sns.jointplot(
    x="carat", 
    y="price",
    color="orange",
    hue='clarity',
    kind="scatter",
    palette="viridis",
    
    data=diamonds.query("carat <= 1 and price<6000")
)
```
