```python
import seaborn as sns
import pandas as pd
```

```python
diamonds = pd.read_csv("../Data/Diamonds Prices2022.csv")
```

```python
sns.boxplot(
    x="clarity",
    y="price",
    color="orange",
    data=diamonds
)
```

```python
sns.violinplot(y="clarity", x="price", palette="husl", data=diamonds)
```
