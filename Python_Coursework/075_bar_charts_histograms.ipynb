```python
import pandas as pd
import seaborn as sns
```

```python
diamonds = pd.read_csv("../Data/Diamonds Prices2022.csv")

diamonds.head()
```

```python
sns.barplot(x="cut", y="price", data=diamonds)
```

```python
sns.barplot(x="price", y="cut", ci=None, data=diamonds)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 8))

sns.barplot(
    y="cut",
    x="price",
    hue="clarity",
    data=diamonds,
    palette="husl"
)


ax.legend(bbox_to_anchor=(1.2, 1))
```

```python
sns.histplot(
    x="price", 
    hue="clarity",
    data=diamonds.query("clarity in ['I1', 'IF']"),
    palette="husl",
    stat='percent',
    bins=10, 
    # binwidth=1000,
    kde=True
)
```

```python
sns.histplot(
    x="price",
    y="clarity",
    data=diamonds,
    cbar=True
)
```

```python
sns.scatterplot(
    x="x", 
    y="price",
    data=diamonds,
    hue='carat',
    palette="husl",
    size="carat"
)
```
